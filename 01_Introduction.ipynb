## **1. What Is Linear Algebra and Why Learn It?**

Linear algebra has a rich history that stretches back to the 17th century in the West, and even earlier in China. Matrices — the central objects of linear algebra — were originally introduced by Descartes as a compact notation for storing geometric coordinates, and later extended by Gauss to represent systems of equations. Throughout the 20th century, these ideas expanded to support multivariate mathematics including calculus, differential equations, physics, and economics.

But the critical turning point came with the rise of computing. Cohen draws a sharp distinction between traditional linear algebra (abstract, proof-based, focused on infinite-dimensional vector spaces) and modern linear algebra (computational, application-oriented, focused on data science, machine learning, and AI). Modern computers are *extremely* efficient at matrix operations — and because nearly every algorithm implemented on a computer ultimately rests on some form of matrix mathematics, linear algebra has become the universal language of computation.

Cohen frames the question "Should you learn linear algebra?" directly: if you want to *understand* algorithms (not just apply them blindly), or if you want to *develop* and *adapt* computational methods for new problems, then yes — and you should learn the *modern* version of it.

This distinction matters because traditional textbooks tend to avoid numerical examples, expect derivations of difficult proofs, and cover topics with little relevance to practical implementation. Cohen's book is an explicit rejection of that approach in favor of intuition through code.

## **4. Mathematical Proofs Versus Intuition from Coding**

This is perhaps the most intellectually interesting part of Chapter 1. Cohen identifies two complementary ways of understanding mathematics:

1. Rigorous proofs — sequences of logical statements that establish mathematical truth with certainty, but often without yielding intuition.
2. Visualizations and examples — concrete, numerical demonstrations that build lasting intuition but risk overgeneralizing from specific cases.

Cohen introduces a third approach he calls "soft proofs" or "mathematical intuition from coding." The idea is elegant: you assume that Python (and libraries like NumPy and SciPy) correctly implement low-level number crunching, while you explore the *principles* by running many numerical experiments. Each successful test reinforces your understanding of a mathematical property without requiring a formal derivation.

Below is Cohen's example of a "soft proof" for the commutative property of multiplication (`a × b = b × a`):

```python
import numpy as np

a = np.random.randn()
b = np.random.randn()
a*b - b*a
```

Cohen is clear that intuition from code is *not* a substitute for rigorous proof — but it is a powerful pedagogical tool, especially for learners who don't have an advanced mathematics background. The principle of soft-proving will be used throughout the entire book.
